# Convert Date Time Format

```python
ndf["date_added"]=pd.to_datetime(ndf['date_added'])
ndf['day_added']=ndf['date_added'].dt.day
ndf['year_added']=ndf['date_added'].dt.year
ndf['month_added']=ndf['date_added'].dt.month
ndf['year_added'].astype(int);
ndf['day_added'].astype(int);
```

# Data Visualization

```python
#Bar graph for netflix copied data set after cleaning "ndf"
```

```python
sb.barplot(x=ndf['type'].value_counts().index, y=ndf['type'].value_counts())
plt.title('TYPE')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
#Bar graph for netflix data set before cleaning "netflix"
```

```python
sb.barplot(x=netflix['type'].value_counts().index,y=netflix['type'].value_counts())
plt.title('TYPE')
plt.xlabel('Type')
plt.ylabel('count')
plt.gcf().set_size_inches(10,10)
plt.show()
```

```python
sb.barplot(x=ndf['rating'].value_counts().index, y=ndf['rating'].value_counts())
plt.title('Rating of ndf dataset')
plt.xlabel('rating')
plt.ylabel('count')
fig=plt.gcf()
fig.set_size_inches(13,13)
plt.show()
```

```python
sb.countplot(x='rating' ,hue='type',data=ndf)
plt.title('Relation between Type and Rating')
plt.gcf().set_size_inches(13,13)
plt.show()
```

```python
size=ndf['type'].value_counts()
labels=['Movie','TV Shows']
colors = plt.cm.Wistia(np.linspace(0,1,2))#colors=['yellow','green']
explode = [0, 0.1]


plt.pie(size,labels=labels,colors=colors,explode=explode,shadow=True,startangle=90, autopct='%1.1f%%')
plt.title('Pie-chart for types distribution',fontsize=10)
plt.show()
```

```python
#labels: A list containing the labels 'Movie' and 'TV show' for the pie chart.

#size: The count of 'Movie' and 'TV show' occurrences in the 'type' column of the 'netflix' DataFrame. It is used to represent the sizes of the slices in the pie chart.

#colors: Using the plt.cm.Wistia() colormap to generate two colors for the slices. The np.linspace(0, 1, 2) function creates two evenly spaced numbers between 0 and 1, which will be used to get two distinct colors from the Wistia colormap.

#explode: A list indicating how much to separate the slices from the center of the pie. Here, we set the second slice ('TV show') to be slightly separated by using a value of 0.1.

#plt.rcParams['figure.figsize'] = (9, 9): Setting the figure size to (9, 9) to make the pie chart display in a square aspect ratio.

#plt.pie(): Creating the pie chart using the 'size', 'labels', 'colors', 'explode', 'shadow', and 'startangle' parameters
```

```python
ndf['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.gcf().set_size_inches(10,108)
plt.title('Pie-chart for rating')
plt.show()
```
